# While Loops

```python
# Infinity Loop
while True:
    print("Hello World")
```

```python
# With condition
message = 'spam'
while message != '':
    print(message, end="\n")
    message = message [1:]
```

```python
message = 'spam'
while message != '':
    print(message, end="\n")
    message = message [:-1]
```

```python
a = 0
b = 11

while a < b: 
    print(a, end=' ')
    a += 1
```

```python
# Break
while True: 
    name = input('Enter Name : ')
    if name == 'stop' : 
        break
    age = input('Enter Age : ')
    print ("\n Hello", name, '\n Umur', age)
```
